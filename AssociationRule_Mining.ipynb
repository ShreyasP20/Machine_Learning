```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
data = pd.read_excel('/content/drive/My Drive/Colab Notebooks/Online Retail.xlsx')
data.head()
```

```python
data.Country.unique()
```

```python
data['Description'] = data['Description'].str.strip()
data.dropna(axis = 0, subset = ['InvoiceNo'], inplace = True)
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]
data.head()
```

```python
basket1_France = (data[data['Country'] == "France"]
        .groupby(['InvoiceNo', 'Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
basket1_France.head()
```

```python
basket1_UK = (data[data['Country'] == "United Kingdom"]
        .groupby(['InvoiceNo', 'Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))


basket1_Por = (data[data['Country'] == "Portugal"]
        .groupby(['InvoiceNo', 'Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))

basket1_Sweden = (data[data['Country'] == "Sweden"]
        .groupby(['InvoiceNo', 'Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
```

```python
def hot_encode1(P):
    if(P<= 0):
        return 0
    if(P>= 1):
        return 1
```

```python
basket1_encoded = basket1_France.applymap(hot_encode1)
basket1_France = basket1_encoded

basket1_encoded = basket1_UK.applymap(hot_encode1)
basket1_UK = basket1_encoded

basket1_encoded = basket1_Por.applymap(hot_encode1)
basket1_Por = basket1_encoded

basket1_encoded = basket1_Sweden.applymap(hot_encode1)
basket1_Sweden = basket1_encoded
```

```python
frq_items1 = apriori(basket1_France, min_support = 0.05, use_colnames = True)

rules1 = association_rules(frq_items1, metric = "lift", min_threshold = 1)
rules1 = rules1.sort_values(['confidence', 'lift'], ascending = [False, False])
```

```python
with pd.option_context('display.max_colwidth', 800):
  print(rules1)
```
