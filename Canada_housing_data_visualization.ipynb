```python
fig = px.scatter_mapbox(df, lat='Latitude', lon='Longitude', color='Price',
                        size='Price',
                        color_continuous_scale='ice',
                        size_max=10,
                        mapbox_style='carto-positron',
                        title='Interactive Map with Price Visualization',
                        hover_name='New_Address',
                        hover_data={'Latitude': True, 'Longitude': True, 'Price': True})

fig.show()
```
